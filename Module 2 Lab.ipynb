```python
# Name: John Nelson II
# File name: Module 2 Lab.ipynb
# Brief desc: My app tests if the students GPA is qualifiable for the honor roll or the deans list. If they are, it will let them know. If they aren't, it will let them know.


print("Hello, welcome to your own personal student quality test, we will see if you are qualified fo the Dean's list or the the Honor Roll!")
print("Please submit the information below:")

for _ in range(5):
    last_name = input("What is your last name: ")
    if last_name == "ZZZ":
        print("Quitting")
        break

    first_name = input("What is your first name: ")
    GPA = float(input("What is your GPA: "))

    if GPA >= 3.5:
        print("Thank you,", first_name, last_name, "You are qualified for the Dean's List.")
    elif GPA >= 3.25:
        print("Thank you,", first_name, last_name, "You are qualified for the Honor Roll.")
    else:
        print("Thank you,", first_name, last_name, "You did not qualify this time.")
    
    print()  # adds a blank line between students


```
